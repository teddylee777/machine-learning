```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.callbacks import ModelCheckpoint, ReduceLROnPlateau
from keras import regularizers
from keras.optimizers import Adadelta, Adam
from keras import backend as K
```

```python
def draw_dist(arr):
    length = arr.shape[1]
    fig, axes = plt.subplots(4, 1, figsize=(8, 12))
    for i in range(length):
        sns.distplot(arr[:, i], ax=axes[i])
```

```python
# best model so far
def build_model_v1():
    model = Sequential()
    model.add(Dense(units=180, activation='relu', 
                    input_dim=226,
                    kernel_initializer='uniform',
                    kernel_regularizer=regularizers.l2(0.01),
                   ))
    model.add(Dense(units=180, activation='relu'))
    model.add(Dense(units=180, activation='relu'))
    model.add(Dense(units=160, activation='relu'))
    model.add(Dense(units=160, activation='relu'))
    model.add(Dense(units=160, activation='relu'))
    model.add(Dense(units=160, activation='relu'))
    model.add(Dense(units=80, activation='relu'))
    model.add(Dense(units=80, activation='relu'))
    model.add(Dense(units=80, activation='relu'))
    model.add(Dense(units=40, activation='relu'))
    model.add(Dense(units=40, activation='relu'))
    model.add(Dense(units=40, activation='relu'))
    model.add(Dense(units=4, kernel_initializer='normal', activation='linear'))
    return model
```

```python
def build_model_v2():
    model = Sequential()
    model.add(Dense(units=180, activation='relu', 
                    input_dim=226,
                    kernel_initializer='uniform',
                   ))
    model.add(Dense(units=180, activation='relu'))
    model.add(Dense(units=180, activation='relu'))
    model.add(Dense(units=160, activation='relu'))
    model.add(Dense(units=160, activation='relu'))
    model.add(Dense(units=160, activation='relu'))
    model.add(Dense(units=160, activation='relu'))
    model.add(Dense(units=120, activation='relu'))
    model.add(Dense(units=120, activation='relu'))
    model.add(Dense(units=100, activation='relu'))
    model.add(Dense(units=80, activation='relu'))
    model.add(Dense(units=60, activation='relu'))
    model.add(Dense(units=40, activation='relu'))
    model.add(Dense(units=2,activation='linear'))
    return model
```

```python
def build_model_v5():
    model = Sequential()
    model.add(Dense(units=180, activation='relu', 
                    input_dim=226,
                    kernel_initializer='normal',
                    kernel_regularizer=regularizers.l2(0.01),
                   ))
    model.add(Dense(units=150,
                    kernel_initializer='normal', 
                    kernel_regularizer=regularizers.l2(0.01),
                    activation='relu'))
    model.add(Dense(units=120,
                    kernel_initializer='normal', 
                    kernel_regularizer=regularizers.l2(0.01),
                    activation='relu'))
    model.add(Dense(units=90,
                    kernel_initializer='normal', 
                    kernel_regularizer=regularizers.l2(0.01),
                    activation='relu'))
    model.add(Dense(units=60,
                    kernel_initializer='normal', 
                    kernel_regularizer=regularizers.l2(0.01),
                    activation='relu'))
    model.add(Dense(units=30,
                    kernel_initializer='normal', 
                    kernel_regularizer=regularizers.l2(0.01),
                    activation='relu'))
    model.add(Dense(units=15,
                    kernel_initializer='normal', 
                    kernel_regularizer=regularizers.l2(0.01),
                    activation='relu'))
        
    model.add(Dense(units=4, kernel_initializer='normal', activation='linear'))
    return model
```

```python
# next best loss so far 1.3999
def my_loss(y_true, y_pred):
    target_mean = 155.0
    target_std = 86.554468
    
    y_true_f = K.flatten(y_true) * 100
    y_pred_f = K.flatten(y_pred) * 100
    
    total_loss = K.mean(K.abs(y_true_f - y_pred_f)) * 5
    diff_mean = K.mean(K.abs(y_true_f - target_mean)) * 3
    diff_std = K.mean(K.abs(K.std(y_true_f) - target_std)) * 3
    return total_loss * diff_mean + diff_std
```

```python
def my_metric(y_true, y_pred):
    y_true_f = K.flatten(y_true)
    y_pred_f = K.flatten(y_pred)
    
    return K.mean(K.abs(y_true_f - y_pred_f)) * 100
```
